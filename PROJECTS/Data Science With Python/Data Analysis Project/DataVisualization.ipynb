```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("churn_cleaned.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='country', data=df)
plt.title("Country Distribution")
plt.show()
```

```python
plt.hist(df['age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df['balance'], bins=20)
plt.title("Balance Distribution")
plt.xlabel("Balance")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df['credit_score'], bins=20)
plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['age'], df['balance'])
plt.title("Age vs Balance")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.show()
```

```python
plt.scatter(df['age'], df['balance'])
plt.title("Age vs Balance")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.show()
```

```python
sns.countplot(x='gender', hue='churn', data=df)
plt.title("Churn vs Gender")
plt.show()
```

```python
sns.countplot(x='country', hue='churn', data=df)
plt.title("Churn vs Country")
plt.show()
```

```python
sns.countplot(x='active_member', hue='churn', data=df)
plt.title("Churn vs Active Members")
plt.show()
```

```python
sns.countplot(x='products_number', hue='churn', data=df)
plt.title("Churn vs Products Number")
plt.show()
```

```python
sns.boxplot(x='churn', y='balance', data=df)
plt.title("Balance vs Churn")
plt.show()
```

```python
sns.boxplot(x='churn', y='age', data=df)
plt.title("Age vs Churn")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

corr = numeric_df.corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
