```python
df.drop('customer_id',axis=1,inplace=True)
```

```python
df['credit_score'] = pd.to_numeric(df['credit_score'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')
df['balance'] = pd.to_numeric(df['balance'], errors='coerce')
df['products_number'] = pd.to_numeric(df['products_number'], errors='coerce')
```

```python
df['credit_score'].fillna(df['credit_score'].mean(), inplace=True)
df['age'].fillna(df['age'].mean(), inplace=True)
df['tenure'].fillna(df['tenure'].mean(), inplace=True)
df['balance'].fillna(df['balance'].mean(), inplace=True)
df['products_number'].fillna(df['products_number'].mode()[0], inplace=True)
```

```python
df['credit_score'] = df['credit_score'].round().astype(int)
```

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

```python
df['gender'] = df['gender'].astype('category')
df['country'] = df['country'].astype('category')
```

```python
df['gender'].unique()
```

```python
df['gender'] = df['gender'].astype(str)
```

```python
df['gender'] = df['gender'].replace(r'^\s*$', np.nan, regex=True)
```

```python
df['gender'] = df['gender'].str.strip().str.capitalize()
```

```python
df['gender'] = df['gender'].astype('category')
```

```python
df['gender'].unique()
```

```python
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

```python
df = df[(df['age'] >= 18) & (df['age'] <= 75)]
```
